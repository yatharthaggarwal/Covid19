```python
def fetchplotdistrict(state):
    print(state)
    print()
    l1 = []
    for i in range(df1.shape[0]):
        if df1['State'][i] == state:
            l1.append(df1['District'][i])
    print(l1)
    
    l2 = []
    print()
    for i in range(zones.shape[0]):
        if zones['State'][i] == state:
            l2.append(zones['District'][i])
    print(l2)        
    print(list(set(l1).difference(set(l2))))
    print(list(set(l2).difference(set(l1))))
```

```python
import json
from bokeh.io import show
from bokeh.io import output_file, save
from bokeh.models import (CDSView, ColorBar, ColumnDataSource,
                          CustomJS, CustomJSFilter, 
                          GeoJSONDataSource, HoverTool,
                          CategoricalColorMapper, LinearColorMapper, Slider)
from bokeh.layouts import column, row, widgetbox
from bokeh.plotting import figure
```

```python
geosource = GeoJSONDataSource(geojson = newdf.to_json())
```

```python
# Define color palettes
palette = ['red', 'orange', 'green']

# reverse order of colors so higher values have darker colors
# palette = palette[::-1] 

# Instantiate LinearColorMapper that linearly maps numbers in a range, into a sequence of colors.
color_mapper = CategoricalColorMapper(palette = palette, factors = ['Red', 'Orange', 'Green'])
```

```python
# Create figure object.
p = figure(title = 'Red, Green and Orange zones of Covid19: District-wise distribution',
           plot_height = 700 ,
           plot_width = 650, 
           toolbar_location = 'right',
           tools = "pan, wheel_zoom, box_zoom, reset")
p.title.text_font_size = '12pt'
p.title.align = 'center'

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None

# p.xaxis.visible = False
# p.yaxis.visible = False

# Add patch renderer to figure.
states = p.patches('xs','ys', source = geosource,
                   fill_color = {'field' :'Zone',
                                 'transform' : color_mapper},
                   line_color = 'black',
                   line_width = 0.25, 
                   fill_alpha = 1)
# Create hover tool
p.add_tools(HoverTool(renderers = [states],
                      tooltips = [('District','@District'),
                                  ('Zone','@Zone'),
                                  ('Confirmed','@Confirmed'),
                                  ('Active','@Active'),
                                  ('Recovered','@Recovered'),
                                  ('Deaths','@Deceased')
                                 ]))

output_file('plot.html', mode='inline')
save(p)
```
